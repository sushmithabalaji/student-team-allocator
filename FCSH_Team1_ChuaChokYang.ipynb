# Python Code
The final Python code compiles all the functions into one complete program. Each part of the code handles a specific task, from reading and cleaning the CSV data to forming and balancing the teams. We used modular programming to keep the code organized, where every function serves a clear purpose and can be tested independently.
We made use of pattern recognition and algorithmic design. Recognising patterns in how students could be distributed fairly and turned those patterns into clear logic within the code. For example, alternating between male and female students and assigning students to the team with the lowest total CGPA would naturally balance both gender and performance.  also used algorithmic design to create loops and conditional checks that maintained fairness and diversity across all tutorial groups.

```python
# sorting functions so 10 doesn't come after 1
import re
def sort_key(s):
    return [int(text) if text.isdigit() else text.lower() for text in re.split(r'(\d+)', s)]

def sort_dict(dict):
    return {key: dict[key] for key in sorted(dict.keys(), key=sort_key)}
```

The function splits each string into separate parts of letters and numbers, converting any numeric parts into integers and change the text parts to lowercase, producing a list that Python can compare correctly when sorting. The second function then takes a dictionary, sorts its keys and creates a new dictionary in that order. 

```python
def read_csv(filename):
    data = []  #storing values in a dictionary
    try:
        file = open(filename, "r")
        lines = file.readlines()
        file.close()

        if not lines:
            print("CSV file is empty")
            return data, []

        headers = [h.strip() for h in lines[0].strip().split(",")] #cleans up header name in dictionary

        for i in range(1, len(lines)):
            line = lines[i].strip()
            parts = [p.strip() for p in line.split(",")] #splits the line into different columns

            record = {
                "TutorialGroup": parts[0],
                 "StudentID": parts[1],
                 "School": parts[2],
                 "Name": parts[3],
                  "Gender": parts[4],
                  "CGPA": float(parts[5])
            }
            data.append(record)
        return data, headers

    except FileNotFoundError:
        print("file not found:", filename)
        return [], []
```

This function reads and processes student data from a CSV file located in the same file path. Try–except block is used to handle missing files and check if the file is empty before continuing. The first line is treated as the header, while each subsequent line is split into the following, TutorialGroup, StudentID, School, Name, Gender, and CGPA. strip() is used to remove extra spaces, and the CGPA value is converted to a float. The function stores every student as a dictionary and returns a list of these dictionaries along with the header names that the rest of the program can easily process.

```python
def group_by_tutorial(data):
    groups = {}
    for student in data: #looping through each student dictionary in data
        group = student["TutorialGroup"]
        if group not in groups:
            groups[group] = [] #create a new list for a new group
        groups[group].append(student)
    return groups
```

The function that controls this is group_by_tutorial. It first creates an empty dictionary for the sorted data. Using a for loop, it loops through the raw data extracted from the CSV file. For each student, it reads the tutorial group, and if the tutorial group has not been added to the dictionary, it adds a new key of the same name as the tutorial group, with an empty array as its value. The code then appends the student to the correct array using their tutorial group as the key. Once all students have been assigned, the dictionary will be returned.


```python
def sort_by_cgpa(tutorial_groups):
    sorted_groups = {}
    for group, students in tutorial_groups.items(): #loop through each tutorial group and its list of students
        sorted_groups[group]= sorted(students, key=lambda s: s["CGPA"], reverse = True) #sort each group from highest to lowest cgpa
        # Split into 2 groups for male and female, organised by CGPA
        sorted_groups[group] = {
            "Male": [student for student in sorted_groups[group] if student["Gender"] == "Male"],
            "Female": [student for student in sorted_groups[group] if student["Gender"] == "Female"]
        }
    return sorted_groups
```

This function is used to sort students in every tutorial group by descending CGPA. It first initialises an empty list of sorted tutorial groups. Using a For Loop, it iterates through the list of students in each tutorial group and using the built-in sorted() function in Python, it sorts the list of students from highest to lowest CGPA by using the key parameter lambda s: s[“CGPA”] which tells Python to use the students’ CGPA as a sorting criterion, and the reverse=True argument ensures that the highest CGPA values appears first. After sorting the students, the function continues to split the list of students in each tutorial group by their genders, creating two sublists of males and females with each list still in descending CGPA order. It finally returns a dictionary of sorted tutorial groups, where each group contains 2 lists of students split by gender and each list is sorted by descending CGPA

```python
## place each student into team with lowest total CGPA that still has capacity
def make_teams(students, teamsize=5):
    n = len(students["Male"]) + len(students["Female"])
    if n == 0:
        return []
    
    num_teams = n // teamsize
    remainder = n % teamsize
    team_capacity = [teamsize] * num_teams
    for i in range(remainder):
        team_capacity[i] += 1
    teams = [[] for i in range(num_teams)] #list of teams in each tutorial group
    sums = [0.0] * num_teams               #list of sum of cgpa of each team

    for s in range(n):
        available_teams = [i for i in range(num_teams) if len(teams[i]) < team_capacity[i]]
        best_team = min(available_teams, key=lambda i: sums[i]) #chooses the team with the lowest cgpa sum
        
        # sort members into teams based on gender and CGPA, round robin style
        # works by reading the gender of the last member added to a team as well as average CGPA of that team
        if teams[best_team] == [] or teams[best_team][-1]["Gender"] == "Female":
            try:
                teams[best_team].append(students["Male"][0])
                sums[best_team] += students["Male"][0]["CGPA"]
                students["Male"].pop(0)
            except IndexError:
                teams[best_team].append(students["Female"][0])
                sums[best_team] += students["Female"][0]["CGPA"]
                students["Female"].pop(0)
        elif teams[best_team][-1]["Gender"] == "Male":
            try:
                teams[best_team].append(students["Female"][0])
                sums[best_team] += students["Female"][0]["CGPA"]
                students["Female"].pop(0)
            except IndexError:
                teams[best_team].append(students["Male"][0])
                sums[best_team] += students["Male"][0]["CGPA"]
                students["Male"].pop(0)
    return teams
```

This function allocates students into teams by considering the team’s capacity and total CGPA of each team so that every team has almost-equal number of students and there is an even spread of CGPA and genders where possible. It takes in the total number of students in each tutorial group and the minimum size of each team. If there are no students, the function returns an empty list of teams. It then calculates the number of teams and assigns a capacity to each team, with leftover students, from an uneven division, distributed one by one to the first few teams. In the assignment process, the algorithm repeatedly selects the best team to assign the student to which is the team with available capacity and lowest CGPA sum to maintain balance in academic performance. Within each team, the algorithm tries to alternate between male and female students by checking the gender of the last member added and adding the opposite gender next, falling back to the available gender if one list empties out. Once there are no more students to assign, the function then returns the list of teams in each tutorial group.

```python
# Count number of unique schools
def school_diversity(team):
    schools = []
    # Loop through each student and add their school to the list
    for student in team:
        schools.append(student["School"])
    # Convert the list to a set to remove duplicates
    unique_schools = set(schools)
    # Return the number of unique schools
    return len(unique_schools)

```

The purpose of school_diversity() is to calculate how many unique schools are present in a team.
How it works: 
Start by initialising an empty list to store the schools of all students in the team.
The function then loops through each student and appends their “School” value to the list.
We then used Python’s built-in set() function to automatically remove duplicates.
A set in Python is a built-in data type represented by curly braces {} that stores a collection of unique items. 
For example if 2 students are from CCDS and 3 are from EEE, the set function will give a set of {“CCDS ”, “EEE”}.
Why use set() instead of a loop?
Using a loop requires extra conditional code to manually check for duplicates whereas the set() function makes it simpler, faster, and more efficient, automatically ensuring each school is only counted once.
Finally, the length of the set obtained using the len() function gives us the number of distinct schools in that team, which becomes the team’s diversity score and returns that value.

```python
# Checks each pair of teams within the same tutorial group and swaps students of same gender if it increases school diversity without affecting CGPA too much
def balance_school(teams, cgpa_tolerance=0.2):
    improved = True
    while improved:
        improved = False
        # Loop through all pairs of teams i < j to avoid repeats
        for i in range(len(teams)):
            for j in range(i + 1, len(teams)):
                # Loop through all students in team i
                for a in range(len(teams[i])):
                    # Loop through all students in team j
                    for b in range(len(teams[j])):
                        s1 = teams[i][a] # student a from team i
                        s2 = teams[j][b] # student b from team j

                        # Swap only if same gender
                        if s1["Gender"] != s2["Gender"]:
                            continue

                        # Swap only if CGPA difference small to maintain balance
                        if abs(s1["CGPA"] - s2["CGPA"]) > cgpa_tolerance:
                            continue

                        # Measure diversity before swap
                        curr_div = school_diversity(teams[i]) + school_diversity(teams[j])

                        # Swap
                        teams[i][a], teams[j][b] = s2, s1
                        new_div = school_diversity(teams[i]) + school_diversity(teams[j])

                        # Keep swap if diversity improves
                        if new_div > curr_div:
                            improved = True
                        else:
                            # Revert if no improvement
                            teams[i][a], teams[j][b] = s1, s2
    return teams
```

The purpose of balance_school() is to improve school diversity across all teams after initial team formation through a heuristic (selection-based) swapping process.
How it works: 
The function compares every possible pair of teams within the same tutorial group using 2 for loops with the inner one being an increment of the outer so no repeated teams are compared.
For each pair of teams, it considers a swap of two students, one from each team only if they meet the conditions of being the same gender to preserve gender balance and their CGPA difference is within the given tolerance to preserve the academic balance.
Before and after the swap, it measures the total school diversity using the school_diversity() helper function.
If school diversity increases, the swap is kept and if not, it’s reverted.
This process continues in a loop controlled by a flag called improved which remains True as long as swap has been made until no further improvement can be made

```python
#loop make teams for all groups
def make_teams_for_all_groups(sorted_groups, teamsize):
    all_teams = {}
    for group, students in sorted_groups.items():
        all_teams[group] = make_teams(students, teamsize=teamsize)
        all_teams[group] = balance_school(all_teams[group])
    return all_teams
```

This function takes in data that has all tutorial groups sorted by CGPA and gender. It first creates an empty dictionary before iterating through each tutorial group and passing students from that tutorial group to the make_teams function, which sorts teams based on their gender and CGPA. This data is then added to the dictionary as a value to a key of the same name as the tutorial group. The data is then passed to balance_school to balance the teams by school. The function then returns the balanced teams grouped by tutorial group.


```python
def save_teams_to_csv(teams_by_group, output_filename="FCSH_Team1_ChuaChokYang.csv"):
    try:
        with open(output_filename, "w") as file:
            file.write("TutorialGroup,StudentID,School,Name,Gender,CGPA,TeamAssigned\n")
            for group in teams_by_group:
                team_number = 1
                for team in teams_by_group[group]:
                    for student in team:
                        line = f"{group},{student['StudentID']},{student['School']},{student['Name']},{student['Gender']},{student['CGPA']},{team_number}\n"
                        file.write(line)
                    team_number += 1
        print("File created successfully!")
    except Exception as e:
        print("Error saving file:", e)
```

Initial_data_analysis creates several graphs to analyse the male to female ratios of the current dataset, as well as the CGPA spread and n

```python
import matplotlib.pyplot as plt
from collections import Counter

def plot_gender(all_teams):
    counts = Counter()
    for group, teams in all_teams.items():
        for team in teams:
            males = sum(s["Gender"] == "Male" for s in team)
            females = sum(s["Gender"] == "Female" for s in team)
            counts[f"{females}F-{males}M"] += 1

    if not counts:
        print("No gender data found.")
        return
    my_labels=[f"{k} ({v} teams)" for k, v in counts.items()]
    plt.pie(
        counts.values(),
        labels=my_labels,
        autopct="%1.1f%%",
        pctdistance=1.15,
        labeldistance=None
    )
    plt.legend(title = "Gender composition", loc="upper right", bbox_to_anchor=(1.6, 1))
    plt.title("Overall Gender Composition Across All Teams")
    plt.show()

def plot_cgpa_distribution(all_teams):
    x_positions = []
    y_values = []

   
    for group, teams in all_teams.items():
        for i, team in enumerate(teams, start=1):
            if team:
                avg_cgpa = sum(s["CGPA"] for s in team) / len(team)
                x_positions.append(len(x_positions))  
                y_values.append(avg_cgpa)

    if not y_values:
        print("No CGPA data found.")
        return

    plt.figure(figsize=(10,6))
    plt.scatter(x_positions, y_values,
                color='mediumseagreen', edgecolors='black', alpha=0.8, s=80)
    plt.ylim(3.5, 5)  
    plt.xticks([])   
    plt.xlabel("Teams")
    plt.ylabel("Average CGPA")
    plt.title("Average CGPA per Team")
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()


def plot_school_diversity(all_teams):
    from collections import Counter
    diversity_counts = Counter()

    for group, teams in all_teams.items():
        for team in teams:
            diversity_counts[school_diversity(team)] += 1

    if not diversity_counts:
        print("No diversity data found.")
        return

    
    diversity_levels = sorted(diversity_counts.keys())
    counts = [diversity_counts[d] for d in diversity_levels]

    plt.figure(figsize=(8,5))
    plt.bar(diversity_levels, counts, color='cornflowerblue', edgecolor='black')
    plt.xlabel("Number of Unique Schools in Team")
    plt.ylabel("Number of Teams")
    plt.title("Distribution of Team School Diversity")
    plt.grid(axis='y', alpha=0.3)
    plt.tight_layout()
    plt.show()

```

```python
def main(teamsize): #main programme, just adding this year so ya'll can reference which functions are used
    print(f"Now running with team size of {teamsize}.")
    filename = "./records.csv"
    data, headers = read_csv(filename)
    tutorial_groups = group_by_tutorial(data) #becomes dictionary
    #sort tutorial groups by number
    tutorial_groups = sort_dict(tutorial_groups)
    sorted_tutorial_by_cgpa = sort_by_cgpa(tutorial_groups) #sort each tutorial group by CGPA
    teams_by_group = make_teams_for_all_groups(sorted_tutorial_by_cgpa, teamsize=teamsize) #iterate for all tutorial groups
    save_teams_to_csv(teams_by_group)
    print("Run completed.")
    
    initial_data_analysis(data)
    plot_gender(teams_by_group)
    plot_cgpa_distribution(teams_by_group)
    plot_school_diversity(teams_by_group)


    return tutorial_groups, sorted_tutorial_by_cgpa, teams_by_group
        
```

The main() function begins by reading the student dataset from records.csv using the read_csv() function, then groups students by tutorial using group_by_tutorial(). These groups are sorted in numerical order through sort_dict() and further organized by CGPA and gender using sort_by_cgpa().
The function then calls make_teams_for_all_groups() to form teams within each tutorial group based on the selected team size, saving the completed team allocations to a new CSV file using save_teams_to_csv().
